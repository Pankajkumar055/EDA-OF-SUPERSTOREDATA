```python
#DATA VISUALIZATION WITH THE HELP OF SEABORN AND MATPLOTLIB

#Q1 COUNT OF CATEGORIES BY PLOTING COUNTPLOT
```

```python
ax=sns.countplot(x='Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Q2 TOTAL SALES IN THE COUNTRY REGIION WISE ? SHOWING WITH THE HELP OF BARGRA
```

```python
sle=df.groupby(['Country','Region'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='Country',y='Sales',data=sle,hue='Region')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# sub-categories wise count of the data with the help of graph
```

```python
ax=sns.countplot(x='Sub-Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# THE TOTAL SALES IN THE CITIES AND REGION WISE SHOWING WITH THE HELP OF BARGRAPH
```

```python
sale=df.groupby(['City','Region'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='City',y='Sales',data=sale,hue='Region')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# THE  DISTRIBUTION OF THE CATEGORY AND SUBCATEGORY SHOWING WITH THE HELP OF BARGRAPH
```

```python

plt.bar('Sub-Category','Category',data=df)
plt.show()
```

```python
#THE TOTAL COUNT OF THE SUBCATEGORY OF THE PRODUCTS 
```

```python
df['Sub-Category'].value_counts()
ax=sns.countplot(x=df['Sub-Category'])
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=90)
plt.show()
```

```python
#TOTAL COUNT OF THE PRODUCT SEGMENT WISE 
```

```python
df['Segment'].value_counts()
ax=sns.countplot(x=df['Segment'])
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
#REGION WISE COUNT OF THE PRODUCT
```

```python
df['Region'].value_counts()
ax=sns.countplot(x=df['Region'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#THE TOTAL PROFIT OF THE SUBCATEGORIES OF THE PRODUCT AND SHOWING WITH THE HELP OF BAR GRAPH
```

```python
ax=sns.barplot(x='Sub-Category',y='Profit',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=90)
plt.show()
```

```python
#THE DISCLOSRE OF THE DISCOUNT AND PROFIT WITH RESPECT TO EACH OTHER WITH THE HELP OF LINECHART
```

```python
ax=sns.lineplot(x='Discount',y='Profit',data=df,color='r',label='Discount')
for bars in ax.containers:
    ax.bar_label(bars)
    
```
